# Import Library and Class

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn. linear_model import LinearRegression
```

# Getting the Data

```python
url = "E:\Tugas3.csv"
names = ['coal_prod_change','gdp']
data = pd.read_csv(url, names=names)
```

# Data Description

## Menampilkan 5 baris teratas dari data

```python
print(data.head())
```

```python
print(data.describe())
```

# Showing the Data

```python
print (data)
```

# Drawing the Scatter Plot 

```python
plt.scatter(data['coal_prod_change'], data['gdp'])
plt.xlabel('Coal Prod Change (TWh)')
plt.ylabel('GDP')
plt.title('Scatter Plot Data Hubungan Coal Prod Change (TWh) Serta GDP di Indonesia')
plt.show()
```

# Put the data into variables x and y, and then reshape 

```python
x = data['coal_prod_change'].values.reshape(-1,1)
y = data['gdp'].values.reshape(-1,1)
```

# Create a Model and Fit it 

```python
model = LinearRegression()
model.fit(x,y)
```

# Get the results 

```python
print('a = ',model.intercept_)
print('b = ',model.coef_)
```

# Predict the Data 

```python
y_pred = model.predict([[23]])
print('predicted response:',y_pred)
```
